```python
# Apply scaler() to all the columns except the 'dummy' variables
num_vars = ['Age',  'Years of Experience','Salary']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

```python
df_train.describe()
```

```python
cpy=df_train.copy()
correlation_matrix=cpy.corr()
```

```python
correlation_threshold = 0.8

# Find highly correlated independent variables
correlated_columns = set()
for i in range(len(correlation_matrix.columns) - 1):  # Exclude the last column ('Salary')
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > correlation_threshold:
            correlated_columns.add(correlation_matrix.columns[i])

# Remove highly correlated independent variables
filtered_matrix = correlation_matrix.drop(columns=correlated_columns, index=correlated_columns)

# Print the filtered matrix
filtered_matrix
```

```python
y_train = df_train.pop('Salary')
X_train = df_train
```

```python
y_train
```

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
# lm.fit(X_train, y_train)

#rfe = RFE(lm, 10)
rfe = RFE(lm, n_features_to_select=3)
#running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
X_train.columns
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train_rfe = X_train[col]
```

```python
X_train_rfe
```

```python
import statsmodels.api as sm
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
X_train_rfe
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit() # Running the linear model
results = sm.OLS(y_train, X_train_rfe).fit_regularized(method='elastic_net', alpha=0, L1_wt=1)
coefficients = results.params
coefficients
```

```python
#Let's see the summary of our linear model
print(lm.summary())
```
